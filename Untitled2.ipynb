```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('jamb_exam_results.csv')
```

```python
df['school_type'].value_counts()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.drop('student_id',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
df_full_train,df_test = train_test_split(df,test_size=0.2,random_state=1)
df_train,df_val = train_test_split(df_full_train,test_size=0.25,random_state=1)


df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = np.log1p(df_train.jamb_score.values)
y_val = np.log1p(df_val.jamb_score.values)
y_test = np.log1p(df_test.jamb_score.values)

del df_train['jamb_score']
del df_val['jamb_score']
del df_test['jamb_score']
```

```python
df_train
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
```

```python
features = ['study_hours_per_week', 'attendance_rate',
       'teacher_quality', 'distance_to_school', 'school_type',
       'school_location', 'extra_tutorials', 'access_to_learning_materials',
       'parent_involvement', 'it_knowledge', 'age', 'gender',
       'socioeconomic_status', 'parent_education_level',
       'assignments_completed'
]
```

```python
train_dicts = df_train[features].to_dict(orient='records')
val_dicts = df_val[features].to_dict(orient='records')
```

```python
train_dicts[0]
```

```python
dv = DictVectorizer(sparse=True)
x_train = dv.fit_transform(train_dicts)
x_val = dv.fit_transform(val_dicts)
```

```python
dt = DecisionTreeRegressor(max_depth=1)
dt.fit(x_train, y_train)
```

```python
from sklearn.tree import export_text
names = dv.get_feature_names_out().tolist()
print(export_text(dt, feature_names=names))
```

#### question 3: 80

```python
scores = []
for d in tqdm([10, 15, 20, 25]):
    rf = RandomForestRegressor(n_estimators=0,
                               random_state=1,
                               n_jobs=-1,
                               max_depth=d,
                               warm_start=True)
    for n in tqdm(range(10, 201, 10)):
        rf.n_estimators = n
        rf.fit(x_train, y_train)
        y_pred = rf.predict(x_val)
        score = np.sqrt(mean_squared_error(y_val, y_pred))
        scores.append((n, d, score))

df_scores = pd.DataFrame(scores, columns=['n_estimators', 'max_depth', 'mse'])
```

```python
df_scores
```

```python
for d in[10,15,20,25]:
    df_subset = df_scores[df_scores.max_depth == d]
    plt.plot(df_subset.n_estimators, df_subset.mse,label=d)
plt.legend()
plt.show()
```

```python
def parse_xb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_auc', 'val_auc']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```
